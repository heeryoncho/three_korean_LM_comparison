```python
'''
KoBERT, KR-BERT, KcBERT 실행 결과(총 50 epoch) 중 
val acc가 가장 높은 test acc를 선별하여 
7겹 교차 검증의 test acc의 평균을 구함.
'''

import re

def find_top_val_test(folder, data_name):
    top_val = 0
    top_test = 0
    flag = False
    with open('{}/{}.txt'.format(folder, data_name), 'r') as f:
        lines = f.readlines()
        for line in lines:
            if 'val acc' in line:
                #result = re.findall(r"[-+]?\d*\.\d+|\d+", line)
                val = float(re.findall(r"0.\d+", line)[0])
                if val > top_val:
                    top_val = val
                    flag = True
                    top_test = 0
                    #print("val", top_val)
                elif val == top_val:
                    top_val = val
                    flag = True
                elif val < top_val:
                    flag = False
                #print(flag)
            if ('test acc' in line) and flag:
                test = float(re.findall(r"0.\d+", line)[0])
                if test > top_test:
                    top_test = test
                    #print("test", top_test)

    return top_val, top_test


def result_summary(result_model, train_type):
    print("======================")
    print(result_model)
    print("======================")
    test_list = []
    for i in range(7):
        top_val, top_test = find_top_val_test(result_model, "{}_{}".format(train_type, i))

        print("{}_{}".format(train_type, i))
        print("final top val:", round(top_val, 5))
        print("final top test:", round(top_test, 5))
        print("----------------------")

        test_list.append(top_test)
    #print(test_list)
    avg = sum(test_list) / len(test_list) 

    print(result_model, train_type, "avg:", round(avg, 5), '\n')

result_model = ['result_kobert', 'result_krbert', 'result_kcbert']

train_type = ['job', 'job_economic', 'job_success', 'job_happiness', 'job_success_economic_happiness', 
              'happiness', 'happiness_economic', 'happiness_success', 'happiness_job', 'happiness_success_economic_job', 
              'all']


for m in result_model:
    for tt in train_type:
        result_summary(m, tt)
```
